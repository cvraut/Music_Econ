```python
from json import dumps,loads
import pandas as pd
from collections import defaultdict
```

```python
BB_data = pd.read_csv("./data/Hot Stuff.csv")
BB_data
```

```python
song_art = set()
for i,r in BB_data.iterrows():
    song_art.add((r['Song'],r['Performer']))
Songs_list =  pd.DataFrame(list(song_art),columns=["Song","Performer"])
```

```python
art_2_song = defaultdict(set)
for s,a in song_art:
    art_2_song[a].add(s.lower())
```

```python
len(song_art)
```

```python
art_data = loads(open("artists_scraped.json").read())
art_data2 = loads(open("stripped_art_data.json").read())

art_data = {**art_data,**art_data2}

for art in art_data:
    art_data[art]['song'] = list(map(lambda s: s.lower(),art_data[art]['Song']))
```

```python
art_2_missing_songs = dict()
num_missing_songs = 0
total_songs_searched_for = 0
arts_w_missing_data = set()
found_songs = set()
for art in art_data:
    data = dict()
    data['found'] = {s.lower() for s in art_data[art]['song']} & art_2_song[art]
    for s in data['found']:
        found_songs.add((art.lower(),s.lower()))
    data['excess'] = set(art_data[art]['song']) - art_2_song[art]
    data['missing'] = art_2_song[art] - set(art_data[art]['song'])
    art_2_missing_songs[art] = data
    total_songs_searched_for += len(art_2_song[art])
    num_missing_songs += len(data['missing'])
    if len(data['missing']) > 0:
        arts_w_missing_data.add(art)
print("missing data for {} out of {} songs".format(num_missing_songs,total_songs_searched_for))
print("{} out of {} artists have missing song data".format(len(arts_w_missing_data),len(art_data)))
```

```python
genius_data = loads(open('./Genesis_data.json').read())
m_genius_data = loads(open('./Genesis_missing.json').read())
songs_ind = {}
for art in genius_data:
    for song in genius_data[art]:
        songs_ind[(art.lower(),song.lower())] = "\n".join(genius_data[art][song])
'''
for art in genius_data2:
    for song in genius_data2[art]:
        songs_ind[(art.lower(),song.lower())] = "\n".join(genius_data2[art][song])
'''
missing_songs = []
num_found = 0
for s,a in song_art:
    if (a.lower(),s.lower()) in songs_ind:
        num_found+=1
    else:
        missing_songs.append((a.lower(),s.lower()))
print("Genius found {} songs".format(num_found))
print("Still mising {} songs".format(len(missing_songs)))
```

```python
from Levenshtein import ratio

bag = art_2_missing_songs['Steve Miller']['excess']
needles = art_2_missing_songs['Steve Miller']['missing']
for needle in needles:
    maps = []
    for item in bag:
        maps.append((ratio(needle,item),item))
    print("Needle: '{}' - matches".format(needle))
    print("\n".join(map(lambda x: "{:5.3f},{}".format(x[0],x[1]),sorted(maps,reverse=True)[:10])))
```
