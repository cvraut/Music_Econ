```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import normalize
import bisect
pd.options.display.max_columns = 99
```

```python
nrc = pd.read_csv("../data/nrc_weekly_cat.csv")
nrc = nrc.drop("Unnamed: 0", axis=1)
nrc["timestep"] = pd.to_datetime(nrc["timestep"])
nrc.head()
```

```python
props = ['trust_prop', 'anger_prop', 'fear_prop', 'negative_prop',
       'sadness_prop', 'surprise_prop', 'positive_prop', 'disgust_prop',
       'joy_prop', 'anticipation_prop']
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
for i in range(len(props)):
    print(props[i])
    plt.subplots(1, 2, figsize=(30, 10))
    ax = plt.subplot(1, 2, 1)
    plot_acf(nrc[props[i]], ax = ax, lags=100)
    ax = plt.subplot(1, 2, 2)
    plot_pacf(nrc[props[i]], ax = ax, lags=100)
    plt.show()
```

```python
def detrender(econ, time):
    reg = LinearRegression().fit(np.reshape(time.tolist(), (-1, 1)), econ)
#     return reg.predict(np.reshape(time.tolist(), (-1, 1)))
    return [i-j for i, j in zip(econ, reg.predict(np.reshape(time.tolist(), (-1, 1))))]
```

```python
week_datetime = [pd.to_datetime(i * 604800 * 10**9) for i in range(-2000, 3000)]
week_ids = list(range(-2000, 3000))
grab_week_id = lambda x: week_ids[bisect.bisect_right(week_datetime, x)]-1 if (x - pd.dto_datetime(0)).days < 0 \
    else week_ids[bisect.bisect_left(week_datetime, x)]
nrc["WeekID"] = [grab_week_id(i) for i in nrc["timestep"]]
```

```python
for i in range(len(props)):
    print(props[i])
    data = nrc[props[i]]
    detrended_data = detrender(data, nrc["WeekID"])
    plt.subplots(1, 3, figsize=(40, 10))
    ax = plt.subplot(1, 3, 1)
    plt.plot(nrc["timestep"], detrended_data, label="Detrended")
    plt.plot(nrc["timestep"], data,  label="Original")
    plt.legend()
    ax = plt.subplot(1, 3, 2)
    plot_acf(detrended_data, ax = ax, lags=100)
    ax = plt.subplot(1, 3, 3)
    plot_pacf(detrended_data, ax = ax, lags=100)
    plt.show()
```

```python
for i in range(len(props)):
    print(props[i])
    data = nrc[props[i]]
    detrended_data = [i-j for i, j in zip(data[1:], data[:-1])]
    plt.subplots(1, 3, figsize=(40, 10))
    ax = plt.subplot(1, 3, 1)
    plt.plot(nrc.loc[1:, "timestep"], detrended_data, label="Detrended")
    plt.plot(nrc["timestep"], data,  label="Original")
    plt.legend()
    ax = plt.subplot(1, 3, 2)
    plot_acf(detrended_data, ax = ax, lags=100)
    ax = plt.subplot(1, 3, 3)
    plot_pacf(detrended_data, ax = ax, lags=100)
    plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

for i in range(len(props)):
    print(props[i])
    data = nrc[props[i]]
    detrended_data1 = detrender(data, nrc["WeekID"])
    detrended_data2 = [i-j for i, j in zip(data[1:], data[:-1])]

    print(" > Is the data stationary ?")
    dftest = adfuller(data, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
   
    print("\n > Is the first de-trended data stationary ?")
    dftest = adfuller(detrended_data1, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
   
    print("\n > Is the firt differenced detrended data stationary ?")
    dftest = adfuller(detrended_data2, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("\n\n")
 
```

```python
for i in range(len(props)):
    print(props[i])
    nrc["{}_detrended".format(props[i])] = [np.nan] + [i-j for i, j in zip(data[1:], data[:-1])]
```
