```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import normalize
```

```python
unemp = pd.read_csv("../data/Econ/EconOriginal/unemp.csv")
unemp.head()
```

```python
unemp_rate = list()
unemp_time = list()
for i in range(unemp.shape[0]):
    curr = unemp.loc[i]
    unemp_rate += curr.tolist()[1:]
    unemp_time += [int(curr.Year) + i / 12 for i in range(len(curr) - 1)]
```

```python
unemp = pd.DataFrame()
unemp["MonthID"] = unemp_time
unemp["Unemp"] = unemp_rate
unemp = unemp.dropna()
```

```python
unemp.head()
```

```python
sp = pd.read_csv("../data/Econ/EconOriginal/S&P_month.csv")
sp["Date"] = pd.to_datetime(sp["Date"])
sp["MonthID"] = [i.year + (i.month - 1) / 12 for i in sp["Date"]]
sp = sp[['Date', "MonthID", "Adj Close"]]
sp.columns = ["DATE", "MonthID", "S&P"]
sp.head()
```

```python
f1 = open("../data/Econ/EconOriginal/fnames.txt", 'r')
fnames = [i.strip() for i in f1.readlines()]
f1.close()
```

```python
dname = [i.split(".csv")[0] for i in fnames]
dname
```

```python
ds = [pd.read_csv("../data/Econ/EconOriginal/{}".format(i)) for i in fnames]
```

```python
for i in range(len(fnames)):
    ds[i]["DATE"] = pd.to_datetime(ds[i]["DATE"])
    ds[i].columns = ["DATE",dname[i]]
    ds[i]["MonthID"] = [i.year + (i.month-1) / 12 for i in ds[i]["DATE"]]
    ds[i][dname[i]] = [i if i != '.' else np.nan for i in ds[i][dname[i]]]
    ds[i][dname[i]] = ds[i][dname[i]].astype("float64")
```

```python
ds = [sp, unemp] + ds
dname = ["S&P", "Unemp"] + dname
```

```python
ds[0].head()
```

```python
start_date = sorted([(dname[i], min(ds[i]["MonthID"])) for i in range(len(ds))], key = lambda x: x[1])
start_date
```

```python
end_date = sorted([(dname[i], max(ds[i]["MonthID"])) for i in range(len(ds))], key = lambda x: x[1], reverse=True)
end_date
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
for i in range(len(ds)):
    print(dname[i])
    plt.subplots(1, 2, figsize=(30, 10))
    ax = plt.subplot(1, 2, 1)
    plot_acf(ds[i][dname[i]].interpolate("linear", limit_direction="both"), ax = ax, lags=100)
    ax = plt.subplot(1, 2, 2)
    plot_pacf(ds[i][dname[i]], ax = ax, lags=100)
    plt.show()
```

```python
def detrender(econ, time):
    reg = LinearRegression().fit(np.reshape(time.tolist(), (-1, 1)), econ)
#     return reg.predict(np.reshape(time.tolist(), (-1, 1)))
    return [i-j for i, j in zip(econ, reg.predict(np.reshape(time.tolist(), (-1, 1))))]
```

```python
for i in range(len(ds)):
    print(dname[i])
    data = ds[i][dname[i]].interpolate("linear",limit = 1000, limit_direction="both")
    detrended_data = detrender(data, ds[i]["MonthID"])
    plt.subplots(1, 3, figsize=(40, 10))
    ax = plt.subplot(1, 3, 1)
    plt.plot(ds[i]["MonthID"], detrended_data, label="Detrended")
    plt.plot(ds[i]["MonthID"], data,  label="Original")
    plt.legend()
    ax = plt.subplot(1, 3, 2)
    plot_acf(detrended_data, ax = ax, lags=100)
    ax = plt.subplot(1, 3, 3)
    plot_pacf(detrended_data, ax = ax, lags=100)
    plt.show()
```

```python
for i in range(len(ds)):
    print(dname[i])
    data = ds[i][dname[i]]
    detrended_data = [i-j for i, j in zip(data[1:], data[:-1])]
    time_id = ds[i]["MonthID"].loc[1:]
    plt.subplots(1, 3, figsize=(40, 10))
    ax = plt.subplot(1, 3, 1)
    plt.plot(time_id, detrended_data, label="Detrended")
    plt.plot(time_id, data[1:],  label="Original")
    plt.legend()
    ax = plt.subplot(1, 3, 2)
    plot_acf(detrended_data, ax = ax, lags=100)
    ax = plt.subplot(1, 3, 3)
    plot_pacf(detrended_data, ax = ax, lags=100)
    plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

for i in range(len(ds)):
    print(dname[i])
    data = ds[i][dname[i]].interpolate("linear",limit = 1000, limit_direction="both")
    detrended_data1 = detrender(data, ds[i]["MonthID"])
    detrended_data2 = [i-j for i, j in zip(data[1:], data[:-1])]

    print(" > Is the data stationary ?")
    dftest = adfuller(data, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
   
    print("\n > Is the first de-trended data stationary ?")
    dftest = adfuller(detrended_data1, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
   
    print("\n > Is the firt differenced detrended data stationary ?")
    dftest = adfuller(detrended_data2, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("\n\n")
 
```

```python
for i in range(len(ds)):
    print(dname[i])
    data = ds[i][dname[i]].interpolate("linear",limit = 1000, limit_direction="both")
    detrended_data1 = detrender(data, ds[i]["MonthID"])
    detrended_data2 = [i-j for i, j in zip(data[1:], data[:-1])]
    ds[i]["{}_detrended".format(dname[i])] = [np.nan] + detrended_data2
```
