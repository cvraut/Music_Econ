```python
from json import dumps,loads
import pandas as pd
from collections import defaultdict
```

```python
BB_data = pd.read_csv("./data/Hot Stuff.csv")
BB_data
```

```python
song_art = set()
for r in BB_data.iterrows():
    song_art.add((r['Song'],r['Performer']))
```

```python
art_2_song = defaultdict(set)
for s,a in song_art:
    art_2_song[a].add(s)
```

```python
len(song_art)
```

```python
art_data = loads(open("artists_scraped.json").read())
```

```python
art_2_missing_songs = dict()
num_missing_songs = 0
total_songs_searched_for = 0
arts_w_missing_data = set()
for art in art_data:
    data = dict()
    data['found'] = set(art_data[art]['Song']) & art_2_song[art]
    data['excess'] = set(art_data[art]['Song']) - art_2_song[art]
    data['missing'] = art_2_song[art] - set(art_data[art]['Song'])
    art_2_missing_songs[art] = data
    total_songs_searched_for += len(art_2_song[art])
    num_missing_songs += len(data['missing'])
    if len(data['missing']) > 0:
        arts_w_missing_data.add(art)
print("missing data for {} out of {} songs from the artists found".format(num_missing_songs,total_songs_searched_for))
print("{} out of {} artists have missing song data".format(len(arts_w_missing_data),len(art_data)))
```
